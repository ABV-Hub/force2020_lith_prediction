```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./data/train.csv', sep=';')
```

```python
data.plot(x='NPHI', y='RHOB', c='CALI', colormap='jet', kind='scatter')
plt.xlim(-0.15, 1.2)
plt.ylim(3.95, 0.5)
```

```python
data.plot(x='NPHI', y='RHOB', c='DCAL', colormap='jet', kind='scatter', marker='.')
plt.xlim(-0.15, 1.2)
plt.ylim(3.95, 0.5)
```

DCAL has very high values, this cell quickly calculates another one by subtracting CALI from BS

```python
data['DIFCAL'] = data['CALI'] - data['BS']
```

```python
data.plot(x='NPHI', y='RHOB', c='DIFCAL', colormap='jet', kind='scatter', marker='.')
plt.xlim(-0.15, 1.2)
plt.ylim(3.95, 0.5)
```

```python
data.plot(x='NPHI', y='RHOB', c='DRHO', colormap='jet', kind='scatter', marker='.')
plt.xlim(-0.15, 1.2)
plt.ylim(3.95, 0.5)
```

```python
lithofacies_groups = data.groupby('FORCE_2020_LITHOFACIES_LITHOLOGY')


for name, lith in lithofacies_groups:
    plt.scatter(lith.NPHI, lith.RHOB, marker='.', label=name)
plt.legend()
plt.show()
```
